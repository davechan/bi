```python
#!/usr/bin/env python

import xlsxwriter
import urllib, urllib2
import requests
import lxml.html
import re
import cgi
from BeautifulSoup import BeautifulSoup
from HTMLParser import HTMLParser

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()


# Create an new Excel file and add a worksheet.
workbook = xlsxwriter.Workbook('iDonate.xlsx')
worksheet = workbook.add_worksheet()

# Start from the first cell. Rows and columns are zero indexed.
row = 0
col = 0

headers = ['OrgName', 'OrgLink', 'LogoLink', 'Review', 'Website', 'Description', 'Funding Need']
# Iterate over the data and write it out
for item in (headers):
    worksheet.write(0, col, item)
    col += 1


for page in range(1,23):
    url = "http://social.theidonate.com/ranking?sort=rank&q=&page="+str(page)+"&rank=&demand="
    #url = "http://social.theidonate.com/en/ranking?sort=rank&q=&page=1&rank=&demand="
    
    html = requests.get(url)
    soup = BeautifulSoup(html.content)
        
    for resultorg in soup.findAll('div',{'class':'charity-summary'}):
        # Get name
        orgname = resultorg('span')[-1].string
        print orgname
    
        # Get link
        orglink = resultorg('a')[0]['href']
        #print orglink
        
        logolink = resultorg.findAll('img',{'class':'charity-logo'})
        #print logolink
    
        objective = resultorg.find('p',{'class':'objective'})
        #print objective.string
    
        # Get detail data from orglink
        baseurl = "http://social.theidonate.com"
        urldetail = baseurl + str(orglink)
        html1 = requests.get(urldetail)
        soup1 = BeautifulSoup(html1.content)
    
        review = strip_tags(str(soup1.find('p',{'class':'review'})))
        website = strip_tags(str(soup1.find('p',{'class':'home'})))
        description = strip_tags(str(soup1.find('p',{'class':'description'})))
        demand = soup1.find('div',{'class':'demand'})
        fundneedscore = demand.find('span',{'class':'score'})
        fundneed = float(fundneedscore.string)/20
    
        print review
        print description
        #review_trim = re.compile("[“|”|‘|’]").sub("",review)
        
        data = [str(orgname).decode("utf-8", "replace"), str(orglink).decode("utf-8", "replace"), str(logolink).decode("utf-8", "replace"), str(review).decode("utf-8", "replace"), str(website).decode("utf-8", "replace"), str(description).decode("utf-8", "replace"),fundneed]
        
        row += 1
        print row
        col = 0
        for item in (data):
            worksheet.write(row, col, item)
            col += 1
        
workbook.close()

# get also chinese version and link using Orglink
```
